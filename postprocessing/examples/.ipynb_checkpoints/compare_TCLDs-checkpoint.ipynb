```python
import numpy as np
import matplotlib.pyplot as plt

import lhcTools as lt
import beamlinegraphics as blg

from hisix import *

path = {}
beam = 1
```

## read the hiSixTrack result

```python
fort3      = '../../../fort.3'
twiss      = '../../../twiss_b1.tfs'

path_0TCLD = '../../../runs_0TCLD.0-10um/'
path_1TCLD = '../../../runs_1TCLD.0-10um/'
path_2TCLD = '../../../runs_2TCLD.0-10um/'
```

```python
losses_0TCLD = losses(fort3,twiss)
losses_1TCLD = losses(fort3,twiss)
losses_2TCLD = losses(fort3,twiss)
```

### Read the summary files

```python
for f in ['fort.999','fort.208','fort.66']:
    losses_0TCLD.ReadHiSixSummary(path_0TCLD + f)
    losses_1TCLD.ReadHiSixSummary(path_1TCLD + f)
    losses_2TCLD.ReadHiSixSummary(path_2TCLD + f)
```

```python
losses_0TCLD.finalize()
losses_1TCLD.finalize()
losses_2TCLD.finalize()
```

## Plot the loss maps

### Zoom to IR7

```python
pth = '/home/phermes/Dropbox/codes/madx/160218_beamlinegraphics/twiss_b1.data.thick'
b   = blg.beamlinegraphics(pth)

xmin,xmax     = 19770,20600
ymin,ymax     = 1e-7,3
b.aspectratio = 1.0
b.fontscale   = 0.9
b.textwidth   = 1.1
b.hspace      = 0.1

b.blgFigure(xmin,xmax,subplots=3)

b.ax1.get_xaxis().set_visible(False)

losses_0TCLD.lossmap(b.ax1)
losses_1TCLD.lossmap(b.ax2)
losses_2TCLD.lossmap(b.ax3)


b.ax1.set_ylim(ymin,ymax)
b.ax2.set_ylim(ymin,ymax)
b.ax3.set_ylim(ymin,ymax)

b.ax1.set_ylabel(r'$\eta$ [m$^{-1}$]')
b.ax2.set_ylabel(r'$\eta$ [m$^{-1}$]')
b.ax3.set_ylabel(r'$\eta$ [m$^{-1}$]')

b.ax3.set_xlabel('Position [m]')

# b.ax1.legend(bbox_to_anchor=[1.0,1.25])

b.ax1.text(20500,0.15,'0 TCLDs')
b.ax2.text(20500,0.15,'1 TCLD')
b.ax3.text(20500,0.15,'2 TCLDs')


xlim(xmin,xmax)

plt.savefig('TCLD_comparison_IR7.pgf',bbox_inches='tight')
```

### Full Ring

```python
pth = '/home/phermes/Dropbox/codes/madx/160218_beamlinegraphics/twiss_b1.data.thick'
b   = blg.beamlinegraphics(pth)

xmin,xmax     = 0,lt.clhc
ymin,ymax     = 1e-7,3
b.aspectratio = 1.0
b.fontscale   = 0.9
b.textwidth   = 1.1
b.hspace      = 0.1

b.blgFigure(xmin,xmax,subplots=3)
b.fig.delaxes(b.blg)                                # remove the blg axis

b.ax1.get_xaxis().set_visible(False)

losses_0TCLD.lossmap(b.ax1)
losses_1TCLD.lossmap(b.ax2)
losses_2TCLD.lossmap(b.ax3)


b.ax1.set_ylim(ymin,ymax)
b.ax2.set_ylim(ymin,ymax)
b.ax3.set_ylim(ymin,ymax)

b.ax1.set_ylabel(r'$\eta$ [m$^{-1}$]')
b.ax2.set_ylabel(r'$\eta$ [m$^{-1}$]')
b.ax3.set_ylabel(r'$\eta$ [m$^{-1}$]')

b.ax3.set_xlabel('Position [m]')

b.ax3.legend(loc='upper left')

xlim(xmin,xmax)

plt.savefig('TCLD_comparison_LHC.pgf',bbox_inches='tight')
```
