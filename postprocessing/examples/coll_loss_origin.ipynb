```python
import numpy as np
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
```

```python
collid = 60
```

```python
df0 = pd.DataFrame()
```

```python
rootdir = '../../B1H.RUNS.LEFT//'
n = 0
m = 0
nmax = 2000
mmax = 10000
for dirname in glob.iglob(rootdir+'run_*/'):
    m+=1
    if m==mmax:
        break
    try:
        imp = pd.read_csv(dirname+'fort.209.gz',delim_whitespace=True)
        imp.columns = ['CID','PID']
        imp = imp[imp.CID==collid]
    
        pid = pd.read_csv(dirname+'fort.822.gz',delim_whitespace=True)
        pid.columns = ['PID','PARID','GENCOL','A','Z','E','ANGLE']
    
        newdf = pd.merge(imp, pid, on=['PID'])
    
        df0 = df0.append(newdf)
        n+=1
        if n==nmax:
            break
        print '\r--> Analyzing run', n,
    except:
        pass
```
