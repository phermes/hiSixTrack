## Loss ratio

Check the peak loss ratio for the case of 0 TCLDs and 1 TCLD in the second loss cluster

### Integrated losses

#### First loss cluster

```python
xmin,xmax = 20282.,20350.
```

```python
# extract bins within the selected range
losses_0TCLD.idx = (losses_0TCLD.cbin > xmin) & (losses_0TCLD.cbin < xmax)
losses_1TCLD.idx = (losses_1TCLD.cbin > xmin) & (losses_1TCLD.cbin < xmax)
```

```python
# calculate the integrated losses in the second DS cluster
losses_0TCLD.eta_tot = sum(losses_0TCLD.cold[losses_0TCLD.idx]/losses_0TCLD.normfac)
losses_1TCLD.eta_tot = sum(losses_1TCLD.cold[losses_1TCLD.idx]/losses_1TCLD.normfac)
```

```python
losses_1TCLD.eta_tot/losses_0TCLD.eta_tot
```

99.5% of the losses in the first cluster are removed with the TCLD8

### Peak loss

### First loss cluster

```python
xmin,xmax = 20282.,20350.
losses_0TCLD.idx = (losses_0TCLD.cbin > xmin) & (losses_0TCLD.cbin < xmax)
losses_1TCLD.idx = (losses_1TCLD.cbin > xmin) & (losses_1TCLD.cbin < xmax)
```

```python
peaklossbin0 = losses_0TCLD.cbin[losses_0TCLD.idx][losses_0TCLD.cold[losses_0TCLD.idx]==max(losses_0TCLD.cold[[losses_0TCLD.idx]])][0]
peaklossbin1 = losses_1TCLD.cbin[losses_1TCLD.idx][losses_1TCLD.cold[losses_1TCLD.idx]==max(losses_1TCLD.cold[losses_1TCLD.idx])][0]
```

```python
losses_0TCLD.etamax = losses_0TCLD.cold[losses_0TCLD.cbin==peaklossbin0][0]/losses_0TCLD.normfac
losses_1TCLD.etamax = losses_1TCLD.cold[losses_1TCLD.cbin==peaklossbin1][0]/losses_1TCLD.normfac
```

```python
losses_1TCLD.etamax/losses_0TCLD.etamax
```
