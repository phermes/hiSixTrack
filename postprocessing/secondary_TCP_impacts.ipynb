```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import hisix as hs
import os
```

Read the data

```python
df = hs.hiST_read_batch('../../HLLHC.B1H.0TCLD.0.1umN/',
                        'fort.822.gz',toucfn='hllhc_2tclds001_toucMap.dat.gz',imax=10000)
```

```python
idx = (df.IX>0) & (df.A==208) & (df.Z==82) & (df.ICOLID==30) & (df.EPX<0)
df[idx].EPX.hist(bins=200)
plt.show()
```

```python
%matplotlib inline
idx = (df.IX>0) & (df.A==208) & (df.Z==82) & (df.ICOLID==30) & (df.EPX<0) & (df.ITURN==14)
# df[idx].EPX.hist(bins=200)

plt.plot(df[idx].EPX,df[idx].EE,'.r')
plt.xlabel('exit angle [mrad]')
plt.ylabel('energy [GeV]')

plt.show()
```

```python
idx = (df.IX>0) & (df.A==208) & (df.Z==82) & (df.ICOLID==30) & (df.EPX<0) 

for turn in np.unique(df[idx].ITURN):
    idx = (df.IX>0) & (df.A==208) & (df.Z==82) & (df.ICOLID==30) & (df.EPX<0) & (df.ITURN==turn) 
    if len(df[idx])>100:
        plt.plot(df[idx].EPX,df[idx].EE,'.',label=str(turn))
        plt.xlabel('exit angle [mrad]')
        plt.ylabel('exit energy [GeV]')
    
plt.legend(loc='lower right')
plt.show()
```
